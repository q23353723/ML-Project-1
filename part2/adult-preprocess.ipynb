```python
import pandas as pd
```

```python
headerName = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num',
    'marital-status', 'occupation', 'relationship', 'race',
    'sex', 'capital-gain', 'capital-loss', 'hours-per-week',
    'native-country'
]
pd_train = pd.read_csv("./data/adult.data", index_col=False, names=headerName)
pd_test = pd.read_csv("./data/adult.test", index_col=False, names=headerName)
```

```python
pd_train.head(1)
```

```python
pd_train_filted = pd_train[{"workclass", "marital-status", "occupation", "relationship", "race", "sex", "native-country", "education-num"}]
pd_train_filted.head()
```

```python
max(pd_train_filted["education-num"])
```

```python
marital_status_list = {}
race_list = {}
workclass_list = {}
sex_list = {}
occupation_list = {}
relationship_list = {}
native_country_list = {}
```

```python
marital_status_index = 1
race_index = 1
workclass_index = 1
sex_index = 1
occupation_index = 1
relationship_index = 1
native_country_index = 1
for index, row in pd_train_filted.iterrows():
    if row['marital-status'].strip() not in marital_status_list:
        marital_status_list[row['marital-status'].strip()] = marital_status_index
        marital_status_index += 1
    if row['race'].strip() not in race_list:
        race_list[row['race'].strip()] = race_index
        race_index += 1
    if row['workclass'].strip() not in workclass_list:
        workclass_list[row['workclass'].strip()] = workclass_index
        workclass_index += 1
    if row['sex'].strip() not in sex_list:
        sex_list[row['sex'].strip()] = sex_index
        sex_index += 1
    if row['occupation'].strip() not in occupation_list:
        occupation_list[row['occupation'].strip()] = occupation_index
        occupation_index += 1
    if row['relationship'].strip() not in relationship_list:
        relationship_list[row['relationship'].strip()] = relationship_index
        relationship_index += 1
    if row['native-country'].strip() not in native_country_list:
        native_country_list[row['native-country'].strip()] = native_country_index
        native_country_index += 1
    
```

```python
print(marital_status_list)
print(race_list)
print(workclass_list)
print(sex_list)
print(occupation_list)
print(relationship_list)
print(native_country_list)
```

```python
import numpy as np
for index, row in pd_train_filted.iterrows():
    pd_train_filted.loc[index, 'marital-status'] = marital_status_list[row['marital-status'].strip()]    
    pd_train_filted.loc[index, 'race'] = race_list[row['race'].strip()]
    pd_train_filted.loc[index, 'workclass'] = workclass_list[row['workclass'].strip()]
    pd_train_filted.loc[index, 'sex'] = sex_list[row['sex'].strip()]
    pd_train_filted.loc[index, 'occupation'] = occupation_list[row['occupation'].strip()]
    pd_train_filted.loc[index, 'relationship'] = relationship_list[row['relationship'].strip()]
    pd_train_filted.loc[index, 'native-country'] = native_country_list[row['native-country'].strip()]
```

```python
pd_train_filted.head()
```

```python
pd_train_filted.to_csv("./data/processed_train.csv", index=False)
```
